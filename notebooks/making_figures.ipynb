```python
numbers = [17500, 23000]
fig,axs=plt.subplots(2,5,figsize=(15,6))
for index,number in enumerate(numbers):
    array = io.imread(f"{base_dir}{image_dir}/{number}_image.tif")
    axs[index,0].imshow(array[0][0],cmap='gray')
    axs[index,0].axis('off')
    array = io.imread(f"{base_dir}{image_dir}/{number}_gt.tif")
    axs[index,1].imshow(array[0][0])
    axs[index,1].axis('off')
    array = io.imread(f"{base_dir}{image_dir}/{number}_learnedsegmentation.tif")
    axs[index,2].imshow(array[0][0])
    axs[index,2].axis('off')
    array = io.imread(f"{base_dir}{image_dir}/{number}_distance.tif")
    axs[index,3].imshow(array[0][0])
    axs[index,3].axis('off')
    array = io.imread(f"{base_dir}{image_dir}/{number}_distance_seg.tif")
    axs[index,4].imshow(-1*array[0][0])
    axs[index,4].axis('off')
axs[0,0].set_title("Image")
axs[0,1].set_title("Ground Truth Segmentation")
axs[0,2].set_title("Predicted Binary Segmentation")
axs[0,3].set_title("Predicted Distance Map")
axs[0,4].set_title("Distance Thresholded Binary")
fig.tight_layout()
```
